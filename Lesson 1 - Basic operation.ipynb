# DEFINIZIONE VARIABILI

Salvare delle variabili ci può aiutare a risolvere problemi più complessi.

Si consideri il seguente esempio:

    Un negozio vende delle sacche di ghiaccio a 1.25 dollari e fa il 20% del profitto. Se vende 500 borse di ghiaccio, quanto profitto genera?

```python
500 * (.2 * 1.25)
```

Quindi il negozio genera 125 dollari di profitto. Per rendere il problema più "gestibile" conviene memorizzarsi questi numeri in alcune variabili.

```python
cost_of_ice_bag = 1.25
profit_margin = .2
number_of_bags = 500
```

Calcoliamo quindi il profitto generato e lo printiamo a schermo.

```python
profit_per_bag = cost_of_ice_bag * profit_margin
total_profit = number_of_bags * profit_per_bag
total_profit
```

# COMBINAZIONI DI CONDIZIONI CON OPERATORI LOGICI

Gli operatori logici `and`, `or` e `not` operano sotto condizioni booleane `True` & `False`.
Nello speicifico `and` e `or` operano due condizioni, mentre `not` opera su una singola condizione.

### Operatore `and`
L'operatore `and` ritorna `True` quando entrambe le condizioni sono `True`. ALtrimenti, ritorna `False`.

```python
my_favorite_number > 0 and my_favorite_number <= 3
```

```python
my_favorite_number < 0 and my_favorite_number <= 3
```

```python
my_favorite_number > 0 and my_favorite_number >= 3
```

```python
True and False
```

```python
True and True
```

### Operatore `or`
L'operatore `or` ritorna `True` se almeno una delle due condizioni è `True`. Ritorna invece `False` se entrambe le condizioni sono `False`.

```python
a_neutral_number = 3
a_neutral_number == 3 or my_favorite_number < 0
```

```python
a_neutral_number != 3 or my_favorite_number < 0
```

```python
my_favorite_number < 0 or True
```

```python
False or False
```

### Operatore `not`
L'operatore `not` ritorna `False` se la condizione è `True`, mentre ritorna `True` se la condizione è `False`. 

```python
not a_neutral_number == 3
```

```python
not a_neutral_number < 0
```

```python
not False
```

```python
not True
```

## Combinazione di condizioni

Gli operatori visti precedentemente, possono essere combinati insieme per avere delle condizioni più complesse. Per farlo sfrutteremo le parentesi `(` e `)` per indicare in che ordine verificare le condizioni (ovviamente prima le considizioni nelle parentesi).

```python
(2 > 3 and 4 <= 5) or not (my_favorite_number < 0 and True)
```

Il risultato è `True` poichè la condizione `or` richiede che una delle due condizioni sia `True`. Si noti che la parte `not` restrituisce proprio un `True`.

```python
not (True and 0 < 1) or (False and True)
```

Se non ci sono le parentesi `(` `)`, allora gli operatori logici leggono le condizioni andando da sinistra a destra.

```python
not True and 0 < 1 or False and True
```
